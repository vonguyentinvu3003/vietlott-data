```python
import json
import numpy as np
from collections import defaultdict

file_path = r"C:\Users\OneBim\Desktop\data\vietlott-data\data\power655.jsonl"

# Đọc dữ liệu
draws = []
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        if line.strip():
            obj = json.loads(line)
            draws.append(obj["result"][:-1])  # bỏ số bonus

# Xây dựng ma trận chuyển trạng thái
transition_counts = defaultdict(lambda: defaultdict(int))

for draw in draws:
    for i in range(len(draw)):
        for j in range(len(draw)):
            if i != j:
                transition_counts[draw[i]][draw[j]] += 1

# Chuyển thành xác suất
transition_probs = {}
for a, next_states in transition_counts.items():
    total = sum(next_states.values())
    transition_probs[a] = {b: count / total for b, count in next_states.items()}

# Hàm sinh bộ số dự đoán theo Markov Chain
def generate_prediction(k=6):
    numbers = list(range(1, 56))
    start = np.random.choice(numbers)  # chọn số bắt đầu ngẫu nhiên
    prediction = [start]

    while len(prediction) < k:
        last = prediction[-1]
        if last in transition_probs:
            next_candidates = list(transition_probs[last].keys())
            probs = list(transition_probs[last].values())
            next_num = np.random.choice(next_candidates, p=probs)
        else:
            # fallback: random nếu không có chuyển trạng thái
            next_num = np.random.choice(numbers)

        if next_num not in prediction:
            prediction.append(next_num)

    prediction.sort()
    return prediction

# Sinh thử 3 bộ
for _ in range(3):
    print("Bộ số dự đoán:", generate_prediction())
```

```python
import json
import numpy as np
from collections import defaultdict
import datetime

file_path = r"C:\Users\OneBim\Desktop\data\vietlott-data\data\power655.jsonl"

# Đọc dữ liệu
draws = []
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        if line.strip():
            obj = json.loads(line)
            date = datetime.datetime.strptime(obj["date"], "%Y-%m-%d").weekday()  # 0=Monday, 6=Sunday
            draws.append((date, obj["result"][:-1]))

# Xây dựng ma trận chuyển trạng thái cho từng thứ
transition_probs_by_day = {}

for weekday in range(7):
    transition_counts = defaultdict(lambda: defaultdict(int))
    
    for d, nums in draws:
        if d == weekday:
            for i in range(len(nums)):
                for j in range(len(nums)):
                    if i != j:
                        transition_counts[nums[i]][nums[j]] += 1
    
    # chuyển thành xác suất
    probs = {}
    for a, next_states in transition_counts.items():
        total = sum(next_states.values())
        probs[a] = {b: c / total for b, c in next_states.items()}
    
    transition_probs_by_day[weekday] = probs

# Hàm sinh bộ số dự đoán theo thứ
def generate_prediction_for_day(weekday, k=6):
    probs = transition_probs_by_day.get(weekday, {})
    numbers = list(range(1, 56))
    start = np.random.choice(numbers)  
    prediction = [start]

    while len(prediction) < k:
        last = prediction[-1]
        if last in probs and probs[last]:
            next_candidates = list(probs[last].keys())
            next_probs = list(probs[last].values())
            next_num = np.random.choice(next_candidates, p=next_probs)
        else:
            next_num = np.random.choice(numbers)

        if next_num not in prediction:
            prediction.append(next_num)

    prediction.sort()
    return prediction

# Ví dụ: dự đoán cho thứ 5 (weekday=3)
print("Bộ số dự đoán (thứ 5):", generate_prediction_for_day(3))
```

```python
%pip install tensorflow

import json
import datetime
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split

file_path = r"C:\Users\OneBim\Desktop\data\vietlott-data\data\power655.jsonl"

# 1. Load data
records = []
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        if line.strip():
            obj = json.loads(line)
            date = datetime.datetime.strptime(obj["date"], "%Y-%m-%d")
            result = obj["result"][:-1]  # bỏ số bonus
            records.append((date, result))

# 2. Encode result thành vector 55 chiều (0/1)
def encode_result(result):
    vec = np.zeros(55)
    for num in result:
        vec[num-1] = 1
    return vec

encoded = [encode_result(r) for _, r in records]
encoded = np.array(encoded)

# 3. Tạo sequence cho LSTM
n_steps = 10  # lấy 10 kỳ trước để dự đoán
X, y = [], []
for i in range(len(encoded) - n_steps):
    X.append(encoded[i:i+n_steps])   # 10 kỳ trước
    y.append(encoded[i+n_steps])     # kỳ tiếp theo
X, y = np.array(X), np.array(y)

# 4. Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, shuffle=False)

# 5. Build LSTM model
model = Sequential([
    LSTM(64, activation="tanh", input_shape=(n_steps, 55)),
    Dense(55, activation="sigmoid")  # output = xác suất cho từng số
])
model.compile(optimizer="adam", loss="binary_crossentropy")

# 6. Train
model.fit(X_train, y_train, epochs=20, batch_size=16, validation_split=0.1)

# 7. Predict
y_pred = model.predict(X_test[:1])  # dự đoán kỳ đầu tiên trong test
probs = y_pred[0]
predicted_numbers = sorted(np.argsort(probs)[-6:] + 1)  # top 6 số có xác suất cao nhất

true_result = np.where(y_test[0] == 1)[0] + 1

print("Kết quả thực tế :", true_result)
print("Kết quả dự đoán:", predicted_numbers)
print("Số trùng khớp   :", set(true_result) & set(predicted_numbers))
```
